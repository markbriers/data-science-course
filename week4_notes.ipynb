## Business understanding

Determining a business understanding can be broken into three distinct sections:

1. Business problem and strategic context
2. Impact
3. Constraints

The output of this phase of CRISP-DM is often a report and project plan, which are reviewed and approved by the responsible business owner and wider stakeholder community before the project commences.

### 1. Business problem and strategic context

All projects (data science or otherwise) start with a goal, which is aligned and contextualised within the wider business vision, goals, and values. The data science project goal leads to project-specific objectives and associated success measures. __This is the starting point for all data science projects__ (but is often forgotten).

***
__Example: NHS COVID-19 app (imagined) - a complex (multifaceted) data science project__

The NHS COVID-19 app is part of the wider NHS Test & Trace programme, which has a goal to "break the chains of transmission to enable society to return to a more normal way of life".

The development of digital health infrastructure to achieve this goal is, in itself, a data science project. The business problem is as follows: __can we use personal digital interventions to improve individual and public health?__

The project's __goal__ is to "use data science to promote maximum societal freedom with minimal risk during a public health emergency". The project's basic idea is based around the creation of personalised digital infrastructure (a smartphone app) to satisfy the following (non-SMART) __objectives__:

- Act as a non-pharmaceutical intervention, breaking the chains of transmission by informing and influencing individual behaviour;
- Shape local area health policy decision making through detailed statistical insights;
- Assess the efficacy of personal digital infrastructure for future (health) emergencies.

***

### 2. Impact

Before specifying measures of success, let's understand the general _types of impact_ that we could generate:

- _Understanding_: As a result of the project, there is an increased understanding of issues, methodology and/or process.
- _Attitudinal_: As a result of the project, there is a change in attitude that results in individual, group or others benefiting.
- _Behavioural_: A change in behaviour, such as decision-making, informed by the project.
- _Capacity_: The project leads to new or enhanced capacity (physical, financial, human resources, social capital and connectivity).

By implicity or explicity contextualising our project objectives with respect to these impact types, we are able to derive appropriate success measures.

***
__Example: NHS COVID-19 app (imagined)__

For each of the objectives listed above, we are able to determine project __success measures__:

- Accurate identification of potential disease transmission and quick alerting of contacts to take necessary action _[Behavourial, Capacity]_
- Directors of Publich Health have detailed insights into local population health indicators, improving the timeliness and efficacy of local communications, planning, and interventions _[Behavioural]_
- Prove that the digital intervention was able to reduce case numbers and deaths _[Attitudinal, Understanding]_

***

We are now in a position to determine success measure __Indicators__. The indicators allow us to consider how well the data science project is performing against the objectives and associated success measures. For each objective / success measure pair, we would expect to have a set of indicator defintions as follows:

- Indicator description;
- Frequency of reporting;
- Source location;
- Availability;
- Responsible owner;
- Indicator target;
- Due date.

The definition of indicators are critical to assessing, communicating, and shaping the project's business impact.

Examples of indicators could be related to the discriminative power of a modelling sytem, the reduction in time taken to make a decision, a binary state of quality assurance, and so on.

Now that we have established the data science project goal and objectives, and linked them to the wider business context, we need to ensure that the project minimises the prospect of harm. What harms can data science projects create?

__Bias and Discrimination__

Because they gain their insights from the existing structures and dynamics of the societies they analyse, data-driven technologies can reproduce, reinforce, and amplify the patterns of marginalisation, inequality, and
discrimination that exist in these societies.
Likewise, because many of the features, metrics, and analytic structures of the models that enable data
mining are chosen by their designers, these technologies can potentially replicate their designers’
preconceptions and biases.
Finally, the data samples used to train and test algorithmic systems can often be insufficiently representative
of the populations from which they are drawing inferences. This creates real possibilities of biased and
discriminatory outcomes, because the data being fed into the systems is flawed from the start.

__Denial of Individual Autonomy, Recourse, and Rights__

When citizens are subject to decisions, predictions, or classifications produced by data science systems, situations may
arise where such individuals are unable to hold directly accountable the parties responsible for these outcomes.
Data science can automate cognitive functions that were previously attributable exclusively to accountable human
agents. This can complicate the designation of responsibility in algorithmically generated outcomes, because
the complex and distributed character of the design, production, and implementation processes of AI systems
may make it difficult to pinpoint accountable parties.
In cases of injury or negative consequence, such an accountability gap may harm the autonomy and violate the rights of the affected individuals.

__Non-transparent, Unexplainable, or Unjustifiable Outcomes__

Many machine learning models generate their results by operating on high dimensional correlations that are beyond the interpretive capabilities of human scale reasoning. In these cases, the rationale of algorithmically
produced outcomes that directly affect decision subjects remains opaque to those subjects. While in some use cases, this lack of explainability may be acceptable, in some applications, where the processed data could harbour traces of discrimination, bias, inequity, or unfairness, the opaqueness of the model may be deeply problematic.

__Invasions of Privacy__

Threats to privacy are posed by data science systems both as a result of their design and development processes, and as a result of their deployment. As data science projects are anchored in the structuring and processing of data, the
development of data science technologies will frequently involve the utilisation of personal data. This data is sometimes
captured and extracted without gaining the proper consent of the data subject or is handled in a way that
reveals (or places under risk the revelation of) personal information. On the deployment end, data science systems that target, profile, or nudge data subjects without their knowledge or consent could in some circumstances be interpreted as infringing upon their ability to lead a private life in
which they are able to intentionally manage the transformative effects of the technologies that influence and shape their development. This sort of privacy invasion can consequently harm a person’s more basic right to pursue their goals and life plans free from unchosen influence.

__Isolation and Disintegration of Social Connection__

While the capacity of data science systems to curate individual experiences and to personalise digital services holds the promise of vastly improving consumer life and service delivery, this benefit also comes with potential risks.
Excessive automation, for example, might reduce the need for human-to-human interaction, while algorithmically enabled hyper-personalisation, by limiting our exposure to worldviews different from ours, might polarise social relationships. Well-ordered and cohesive societies are built on relations of trust, empathy, and mutual understanding. As data science technologies become more prevalent, it is important that these relations be preserved.

__Unreliable, Unsafe, or Poor-Quality Outcomes__

Irresponsible data management, negligent design and production processes, and questionable deployment practices can, each in their own ways, lead to the implementation and distribution of data science systems that produce unreliable, unsafe, or poor-quality outcomes. These outcomes can do direct damage to the wellbeing of individual persons and the public welfare. They can also undermine public trust in the responsible use of societally beneficial AI technologies, and they can create harmful inefficiencies by virtue of the dedication of
limited public resources to inefficient or even detrimental data science technologies.

In addition to the ethical and societal considerations listed above, there are practical constraints that also need to be considered. For example, does the project team have the necessary resources (financial, compute, domain experts) available to complete the project? What are the key risks in the project? ### Exploratory Data Analysis

The field of Statistics was originally invented to allow Governments to understand properties of their _populations_ given a _sample_ of data. We should use this framing and terminology when considering the utility of a dataset in the context of the business problem.

Exploratory Data Analysis (EDA) is a statistical phrase governing the "Data Understanding and Data Preparation" parts of the CRISP-DM process. EDA involves graphical and non-graphical exploration of the data. During EDA, we will think about the following:

* Explore the quality of the data collection: Listing the data sources acquired together with their locations, the methods used to acquire them and any problems encountered
* Characterise the distribution of key attributes (for example, the target attribute of a prediction task)
* Produce a description of the data that has been acquired including its format, its quantity (for example, the number of records and fields in each table).
* Produce a description of any other surface features that have been discovered. Evaluate whether the data acquired satisfies the business requirements **(univariate analysis)**
* Characterise the relationships between pairs or small numbers of attributes **(multivariate analysis)**
* Results of simple aggregations
* Properties of significant sub-populations
* Simple statistical analyses, including missing data analysis and anomaly detection

_Structured data_ often arrives in tabular format - that is, can be readily stored and manipulated in a Pandas DataFrame object within Python. The data _fields_ are often stored as columns, and each data point is stored as a row.

_Unstructured data_ often has no discernable format. One of the initial tasks is to devise structure in order to undertake a statistical analysis. We will not cover unstructured data in this course.

__Categorical data__

Categorical data (https://en.wikipedia.org/wiki/Categorical_variable) are variables that take one of a finite number of values (from a set of defined categories).

(From wikipedia) Examples of values that might be represented in a categorical variable:

* The blood type of a person: A, B, AB or O.
* The state that a person lives in.
* The political party that a voter might vote for, e. g. in a European country Christian Democrat, Social Democrat, Green Party, etc.
* The type of a rock: igneous, sedimentary or metamorphic.
* The identity of a particular word (e.g., in a language model): One of V possible choices, for a vocabulary of size V.

We will follow a process of analysing categorical data, inspired by: https://www.datacamp.com/community/tutorials/categorical-data. Let's start by importing the libraries that we need.

```
import pandas as pd
import numpy as np
import copy

import ssl
ssl._create_default_https_context = ssl._create_unverified_context
```

Load the data and display the first few lines:

```
df_flights = pd.read_csv('https://raw.githubusercontent.com/ismayc/pnwflights14/master/data/flights.csv')
df_flights.head()
```

```
print(df_flights.info())
```

We will now convert the 'carrier', 'tailnum', 'origin' and 'dest' fields into Categorical variables (i.e. assign each distinct category with a numerical values) as follows:

```
df_flights['carrier'] = pd.Categorical(df_flights['carrier'])
df_flights['tailnum'] = pd.Categorical(df_flights['tailnum'])
df_flights['origin'] = pd.Categorical(df_flights['origin'])
df_flights['dest'] = pd.Categorical(df_flights['dest'])
print(df_flights.info())
```

We can see that the four variables are now represented in the df_flights DataFrame as Categorical variables. Now let's create a new DataFrame containing the Categorical variables only:

```
cat_df_flights = df_flights.select_dtypes(include=['category']).copy()
cat_df_flights.head()
```

The Categorical python objects contain two properties, categories and codes:

```
print(cat_df_flights["tailnum"].cat.categories)
print("\n---\n")
print(cat_df_flights["tailnum"].cat.codes)
```

_Missing data_: One of the most common data pre-processing steps is to check for null values in the dataset (i.e. missing data). You can get the total number of missing values in the DataFrame by the following:

```
print(cat_df_flights.isnull().sum())
```

There are 248 missing values in the field tailnum, which are represented by the Python object _null_.

You can do a _mode imputation_ for those null values. The function fillna() is handy for such operations.
Note the chaining of method .value_counts() in the code below. This returns the frequency distribution of each category in the feature, and then selecting the top category, which is the mode, with the .index attribute.

```
mode_tailnum = cat_df_flights['tailnum'].value_counts().index[0]
print(mode_tailnum)
cat_df_flights['tailnum'] = cat_df_flights['tailnum'].fillna(mode_tailnum)
print(cat_df_flights.isnull().sum())
```

The above code computes the mode (i.e. the most frequent) 'tailnum' in the dataset (N223AG) and uses this value to populate the dataset.

_Anomaly characterisation_: Now we have _inputed_ missing values, we can view counts across the categorical values:

```
print(cat_df_flights['carrier'].value_counts())
```

We can also produce a plot of these values:

```
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(12,10))
carrier_count = cat_df_flights['carrier'].value_counts()
sns.set(style="darkgrid")
sns.barplot(x=carrier_count.index, y=carrier_count.values, alpha=0.9)
plt.title('Frequency Distribution of Carriers')
plt.ylabel('Number of Occurrences', fontsize=12)
plt.xlabel('Carrier', fontsize=12)
plt.show()
```

Alaska Airways (AS) (encriched via http://www.sterlingtravel.co.uk/airline_list.html) appears to be anomalous - the number of flights from Seattle and Portland is significantly greater than other airlines. This is perhaps to be expected, as https://newsroom.alaskaair.com/cities-served states that "Alaska calls Seattle home", and "Alaska Air Group offers more flights to more destinations from the West Coast than any other airline".
